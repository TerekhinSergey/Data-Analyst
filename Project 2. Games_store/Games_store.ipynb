## Предобработка данных

```python
#приведем названия столбцов к нижнему регистру
df.columns = df.columns.str.lower()
df.columns
```

```python
#проверим на неявные дубликаты
unique_names = ['name', 'platform', 'genre', 'rating']
for i in unique_names:
    display(df[i].unique())
```

```python
#посмотрим на распределение игр по рейтингу
df.rating.value_counts()
```

У нас имеется крайне неравномерное распределение по рейтингам. Чтобы не перегружать последующую визуализацию незначительными по объему рейтингами, приведем данный десяток к ближайшим рейтингам по следующему принципу:
- «EC» («Early childhood» — «Для детей младшего возраста»: Игра подходит для детей от 3 лет) и "K-A" ("Kids to Adults", предыдущий вариант рейтинга «E») к «E» («Everyone» — «Для всех»: Содержание вполне подходит для возрастной категории от 6 лет)
- «AO» («Adults Only 18+» — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет) к  «M» («Mature» — «Для взрослых»: Материалы игры не подходят для подростков младше 17 лет)
- «RP» («Rating Pending» — «Рейтинг ожидается») по имеющимся сейчас данным в интернете к «E10+» («Everyone 10 and older» — «Для всех от 10 лет и старше»)  

Также, в связи с тем, что в дальнейшем исследовании мы будем оценивать влияние возрастного рейтинга на продажи в отдельных регионах, для учета всех данных заменим пропуски на явное их отображение

```python
#напишем функцию для замены рейтинга
def rating_change(rating1, rating2):
    df.loc[df.rating == rating1, 'rating'] = rating2
```

```python
#заменим рейтинг и посмотрим на полученный результат
rating_change('EC', 'E')
rating_change('K-A', 'E')
rating_change('AO', 'M')
rating_change('RP', 'E10+')
df.rating = df.rating.fillna('unknown')
df.rating.value_counts()
```

В столбце с рейтингом игроков имеется аббревиатура TBD (to be determined),обозначающая, что рейтинг в данный момент отсутствует.  
Посмотрим на количество таких строк и можно ли от них избавиться

```python
df[df.user_score == 'tbd']
```

Количество строк значительно, поэтому целесообразнее будет заменить на пропуски, т.к. рейтинг в данный момент нам неизвестен

```python
df.user_score = df.user_score.replace('tbd', np.nan)
```

```python
#посмотрим на две игры без названия
df[df.name.isna()]
```

Определить, что это за игры даже без указания жанра не представляется возможным, а их процент ко всему ДФ пренебрежительно мал, поэтому просто удалим их.

```python
df = df.dropna(subset=['name'])
df.name.isna().sum()
```

```python
#посмотрим на отсутствующие данные в оставшихся столбцах
df[df.year_of_release.isna()]
df[df.critic_score.isna()]
df[df.user_score.isna()]
```

Никакой явной закономерности между отсутствующими данными и определенной платформой или жанром заметить не удалось, поэтому просто оставим эти пропуски в том виде, в каком они имеются сейчас. 

```python
#приведем типы данных к необходимым и проверим полученный результат
df.year_of_release = pd.to_datetime(df.year_of_release, format='%Y')
df.critic_score = df.critic_score.astype('Int8')
df.user_score = df.user_score.astype('float16')
df.info()
```

```python
#для проведения дальнейшего исследования посчитаем суммарные продажи во всех регионах
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
df
```

### Определение актуального периода для анализа

Для ответа на этот вопрос нам необходимо изучить количество выпускаемых игр за каждый год и определить "срок жизни" каждой платформы.

```python
(df.pivot_table(index = 'year_of_release', values = 'name', aggfunc = 'count')                   
   .plot(style='o-', figsize=(15, 8), grid=True, legend=None))
plt.title('Количество выпущенных игр за каждый год', fontsize=20)
plt.xlabel('Год', fontsize=20)
plt.xticks(size = 15)
plt.ylabel('Количество, шт', fontsize=20)
plt.yticks(size = 15) 
plt.annotate('Начало резкого роста', xy=('2000', 400), xytext=('1990', 600),
            arrowprops=dict(facecolor='g'), fontsize=15)
plt.annotate('Начало \nрезкого спада', xy=('2009', 1400), xytext=('2011', 1350),
            arrowprops=dict(facecolor='g'), fontsize=15)
plt.show();
```

Можем увидеть, что количество выпущенных игр постепенно увеличивается. С 1994 года количество игр перевалило за сотню, в 2006 за тысячу, а с 2012 падает почти в два раза и вплоть до 2016 находится в районе 600. Отсюда можно сделать вывод, что наибольшее количество данных у нас имеется в период с 2006 по 2011 и чуть меньшее, но чуть более актуальные данные в период с 2012 по 2016. Снижение количества выпускаемых игры почти в два раза связано, возможно, с развитием индустрии многопользовательских онлайн и мобильных игр, поддерживаемых зачастую десятилетиями.

```python
#посмотрим на изменение продаж по платформам в течение нескольких лет
(df.pivot_table(index = ['platform', 'year_of_release'],
                values = 'total_sales', aggfunc = 'sum')
   .reset_index().tail(60))
```

По данной таблице можно заметить, что за исключением некоторых платформ, как правило, продажи растут вплоть до 3-5 года после выпуска платформы, после чего начинают снижаться, заканчиваясь на 7-10 году. Выберем "топовые" по количеству продаж платформы.

```python
#определим топ10 платформ по продажам
top10 = df.pivot_table(
    index = 'platform', values = 'total_sales', aggfunc = 'sum').sort_values(
    by = 'total_sales', ascending = False).head(10)
top10
```

Посмотрим на распределение продаж по годам этих платформ и визуализируем его.

```python
#напишем функцию для построения графиков
def graph(df, title, xlabel, ylabel):
    df.plot(style='o-', figsize=(20, 10), grid=True)
    plt.title(title, fontsize=20)
    plt.legend(loc='upper left', fontsize='x-large')
    plt.xlabel(xlabel, fontsize=20)
    plt.xticks(size = 15)
    plt.ylabel(ylabel, fontsize=20)
    plt.yticks(size = 15) 
```

```python
df_top10 = df.query('platform in @top10.index')
svod_top10 = df_top10.pivot_table(index = 'platform', columns = 'year_of_release',
                                  values = 'total_sales', aggfunc = 'sum')
graph(svod_top10.T, 'Распределение продаж "топовых" платформ по годам',
      'Год', 'Продажи, млн')
plt.show();
```

По данному графику можно снова заметить, что "срок жизни" платформ составляет примерно 8-10 лет, а новые поколения консолей создаются примерно каждые 6-7 лет. Чтобы сделать более точный прогноз на 2017 год, выберем актуальный период в 2013-2016 года и снова отберем "топовые" платформы. 

```python
#создаем ДФ только с актуальным периодом для анализа
active = df[df.year_of_release >= '2013']
active
```

Определим топовые платформы по актуальному периоду.

```python
top_act = (active.pivot_table(index = 'platform', values = 'total_sales', aggfunc = 'sum')
                 .sort_values(by = 'total_sales', ascending = False).head(10))
display(top_act)
#для анализа оставим 5 платформ с продажами свыше 100млн
top5 = top_act.head(5)
```

Посмотрим на распределение продаж по годам этих платформ и визуализируем его.

```python
top5act = active.query('platform in @top5.index')
svod_top5 = (top5act.pivot_table(index = ['platform', 'year_of_release'],
                                 values = 'total_sales', aggfunc = 'sum')
             .unstack('platform'))
svod_top5.columns = ['3DS', 'PS3', 'PS4', 'X360', 'XOne']
graph(svod_top5, 'Распределение продаж "топовых" платформ по годам',
      'Год', 'Продажи, млн')
plt.show();
```

## Проверка гипотез

### Различаются ли средние пользовательские рейтинги игр на платформах Xbox One и PC

Для проверки этой гипотезы сформируем нулевую и альтернативную гипотезу.  
**Нулевая H0:** средний пользовательский рейтинг игр на Xbox One равен среднему пользовательскому рейтингу игр на PC   
**Альтернативная H1:** средний пользовательский рейтинг игр на Xbox One не равен среднему пользовательскому рейтингу игр на PC 

```python
#для начала отфильтруем наш ДФ по платформам и для каждой посчитаем средней рейтинг
pc = active.query('platform == "PC"').dropna()
x1 = active.query('platform == "XOne"').dropna()
display(pc.shape, x1.shape)
display(pc.user_score.mean())
x1.user_score.mean()
```

```python
alpha = .05 # критический уровень статистической значимости

results = st.ttest_ind(pc.user_score, x1.user_score)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно почти на 17%. Это слишком большая вероятность, чтобы делать вывод о значимом различии между средними рейтингами пользователей игр на PC и Xbox. Мы не использовали дополнительный параметр equal_var = False, потому что у нас достаточно большие выборки, к тому же в обеих измерена одна и та же переменная — нет оснований полагать, что разброс будет сильно отличаться. И действительно - средние рейтинги почти одинаковы.

### Различаются ли средние пользовательские рейтинги игр жанров Action и Sports 

Для проверки этой гипотезы сформируем нулевую и альтернативную гипотезу.  
**Нулевая H0:** средний пользовательский рейтинг игр жанра Action равен среднему пользовательскому рейтингу игр жанра Sports   
**Альтернативная H1:** средний пользовательский рейтинг игр жанра Action не равен среднему пользовательскому рейтингу игр жанра Sports 

```python
#для начала отфильтруем наш ДФ по платформам и для каждой посчитаем средней рейтинг
action = active.query('genre == "Action"').dropna()
sports = active.query('genre == "Sports"').dropna()
display(action.shape, sports.shape)
display(action.user_score.mean())
sports.user_score.mean()
```

```python
alpha = .05 # критический уровень статистической значимости

results = st.ttest_ind(action.user_score, sports.user_score)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Полученное очень маленькое значение p-value говорит о том, что вероятность получить такое или большее различие точно не случайно. Это явно слишком маленькая вероятность, чтобы делать вывод о тождественности между средними рейтингами этих жанров.  И действительно - средние рейтинги отличаются более, чем на единицу.

### Вывод

<div style="border:solid green 2px; padding: 20px">
Нами было выдвинуто две гипотезы: 
    
- средний пользовательский рейтинг игр на Xbox One равен среднему пользовательскому рейтингу игр на PC  
- средний пользовательский рейтинг игр жанра Action равен среднему пользовательскому рейтингу игр жанра Sports  

После проведенных t-тестов мы выяснили, что **первую гипотезу нет оснований отвергать**, а **вторая гипотеза скорее всего всё-таки неверна**.
