# Исследование тарифных планов «Смарт» и «Ультра»

Компания «Мегалайн» — федеральный оператор сотовой связи. Клиентам предлагается два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.
Нам предстоит сделать предварительный анализ тарифов на небольшой выборке клиентов. В нашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год.

**Описание тарифов**  
Тариф **«Смарт»**  
Ежемесячная плата: 550 рублей
Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика
Стоимость услуг сверх тарифного пакета: 1. минута разговора: 3 рубля («Мегалайн» всегда округляет вверх значения минут и мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута); 2. сообщение: 3 рубля; 3. 1 Гб интернет-трафика: 200 рублей.    
Тариф **«Ультра»**  
Ежемесячная плата: 1950 рублей
Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
Стоимость услуг сверх тарифного пакета: 1. минута разговора: 1 рубль; 2. сообщение: 1 рубль; 3. 1 Гб интернет-трафика: 150 рублей.

# Цель проекта.
    
Проанализировать поведение клиентов и сделать вывод — какой тариф более прибыльный для компании - «Смарт» или «Ультра».

# Задачи проекта.
Во время исследования необходимо решить следующие задачи:

1. Изучить исходные данные и провести их предобработку
2. Определить, сколько минут разговора, сообщений и интернет-трафика требуется пользователям каждого тарифа в месяц.
3. Проверить гипотезы, что средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются, а также, что средняя выручка пользователи из Москвы отличается от выручки пользователей из других регионов.

```python
NORM = '\033[0m'  # все атрибуты по умолчанию
BOLD = '\033[1m'  # жирный шрифт (интенсивный цвет)

#Напишем функцию для вывода основной информации о ДФ.
df_list = [df_call, df_inet, df_msg, df_tariff, df_user]
[df_call.name, df_inet.name, df_msg.name, df_tariff.name, df_user.name] = (
    ['Данные о звонках', 'Данные об интернет-трафике', 'Данные о сообщениях',
     'Данные о тарифах', 'Данные о пользователях'])
def full_info(df_list):
    for df in df_list:
        print(BOLD+df.name)
        if len(df.index) > 10:
            print(BOLD+'\n Случайная выборка 10 строк датафрейма:')
            display (df.sample(n=10, random_state=10))
        else:
            print(BOLD+'\n Первые 10 строк датафрейма:')
            display (df.head(10))
        print(BOLD+'\n Типы данных и общая информация: \n'+NORM)
        display(df.info())
        print(BOLD+' Наличие дубликатов: '+NORM, df.duplicated().sum())
        print(BOLD+'\n Наличие пропусков: \n'+NORM)
        display(df.isna().sum())
        print(BOLD+'\n Статистики количественных переменных :')
        display(df.describe())
        print('*'*50)
    return
```

## Предобработка данных

```python
#для оптимизации данных и определения используемого объема памяти создадим функцию
def mem_usage(pandas_obj):
    if isinstance(pandas_obj,pd.DataFrame):
        usage_b = pandas_obj.memory_usage(deep=True).sum()
    usage_mb = usage_b / 1024 ** 2 # преобразуем байты в мегабайты
    return "{:03.2f}".format(usage_mb), 'Mb'
for i in df_list:
    display(mem_usage(i))
```

```python
#поменяем тип данных и посмотрим на используемый объем памяти после этого
df_call.user_id = df_call.user_id.astype('uint16')
df_inet.user_id = df_inet.user_id.astype('uint16')
df_msg.user_id = df_msg.user_id.astype('uint16')
df_call.call_date = pd.to_datetime(df_call.call_date, format='%Y-%m-%dT%H:%M:%S')
df_inet.session_date = pd.to_datetime(df_inet.session_date, format='%Y-%m-%dT%H:%M:%S')
df_msg.message_date = pd.to_datetime(df_msg.message_date, format='%Y-%m-%dT%H:%M:%S')
df_user.reg_date = pd.to_datetime(df_user.reg_date, format='%Y-%m-%dT%H:%M:%S')
df_user.churn_date = pd.to_datetime(df_user.churn_date, format='%Y-%m-%dT%H:%M:%S')
for i in df_list:
    display(mem_usage(i))
```

## Расчёты и добавление результатов в таблицу

### Количество сделанных звонков и израсходованных минут разговора по месяцам

Для того, чтобы посчитать эту информацию для каждого пользователя, для начала необходимо округлить использованные минуты в большую сторону по правилам использования тарифов.

```python
#округлим и проверим результат
df_call.duration = df_call.duration.apply(np.ceil)
df_call
```

```python
#добавим столбец с месяцем звонка
df_call['month'] = df_call['call_date'].dt.month
#построим сводную таблицу по пользователю и месяцу с количеством и длительностью звонков
df_call2 = df_call.pivot_table(index = ['user_id', 'month'], values = 'duration', aggfunc = ['count', 'sum']).reset_index(level = 1)
#переименуем столбцы
df_call2.columns = ['month','amount', 'total_duration']
#и посмотрим на полученный результат
df_call2
```

Получили сводную таблицу с необходимыми нам данными. Переходим к следующему пункту.

### Количество отправленных сообщений по месяцам

Аналогично предыдущему пункту сгруппируем данные и посчитаем необходимую нам информацию.

```python
#добавим столбец с месяцем отправления сообщения
df_msg['month'] = df_msg['message_date'].dt.month
#построим сводную таблицу по пользователю и месяцу с количеством сообщений
df_msg2 = df_msg.pivot_table(index = ['user_id', 'month'], values = 'id', aggfunc = 'count').reset_index(level = 1)
#переименуем столбцы
df_msg2.columns = ['month','amount']
#и посмотрим на полученный результат
df_msg2
```

Получили сводную таблицу с необходимыми нам данными. Переходим к следующему пункту.

### Объем израсходованного интернет-трафика по месяцам

Аналогично предыдущему пункту сгруппируем данные и посчитаем необходимую нам информацию.

```python
#добавим столбец с месяцем отправления сообщения
df_inet['month'] = df_inet['session_date'].dt.month
#построим сводную таблицу по пользователю и месяцу с количеством сообщений
df_inet2 = df_inet.pivot_table(index = ['user_id', 'month'], values = 'mb_used', aggfunc = 'sum').reset_index(level = 1)
#переименуем столбцы
df_msg2.columns = ['month','amount']
#и посмотрим на полученный результат
df_inet2
```

Теперь по условиям тарифа нам необходимо округлить использованные мегабайты в большую сторону

```python
#для этого переведем использованный траффик в гигабайты и затем уже округлим. посмотрим на результат
df_inet2.mb_used = df_inet2.mb_used/1024
df_inet2.mb_used = df_inet2.mb_used.apply(np.ceil)
df_inet2 = df_inet2.rename(columns={'mb_used': 'gb_used'})
df_inet2
```

Получили сводную таблицу с необходимыми нам данными. Переходим к следующему пункту.

### Помесячная выручка с каждого пользователя

Для проведения данных рассчетов нам сначала необходимо объединить полученные сводные таблицы. Для корректного объединения, чтобы избежать потери данных, подготовим "рыбу" со всеми месяцами использования тарифов и всеми пользователями.

```python
#определим крайние даты использования тарифов
df_call.call_date.agg([min, max])
```

```python
#создадим ДФ со всеми месяцами
month = pd.DataFrame(pd.date_range('2018-01-01','2018-12-31',freq='MS'))
month.columns = ['month']
month['month'] = month['month'].dt.month
month
```

```python
#создадим ДФ со всеми пользователями и месяцем начала их использования тарифом
pd.options.mode.chained_assignment = None  # default='warn'
user = df_user[['user_id', 'reg_date']]
user['reg_date'] = user['reg_date'].dt.month
user
```

```python
#создадим общий ключ для объединения таблиц
month['key'] = 1
user['key'] = 1
```

```python
#объединим созданные ДФ по ключу, а затем уберем его
df_new = month.merge(user, on='key')
df_new.drop('key', axis=1, inplace=True)
df_new
```

```python
#усечем наш ДФ на те месяца, когда пользователь еще не начал пользоваться тарифом 
df_new = df_new[df_new['month'] >= df_new['reg_date']]
df_new
```

```python
#проведем присоединение нового ДФ со звонками к нашей "рыбе". Аналогично с сообщениями и интернет-траффиком
df_new=df_new.merge(df_call2, how='left', on=['month','user_id'])
df_new
```

```python
df_new=df_new.merge(df_msg2, how='left', on=['month','user_id'], suffixes=('_call', '_msg'))
df_new
```

```python
df_new=df_new.merge(df_inet2, how='left', on=['month','user_id'])
df_new
```

```python
#для присоединения ДФ с пользовательскими данными преобразуем его и уберем лишнюю информацию
df_user['month'] = df_user['reg_date'].dt.month
df_user2 = df_user[['user_id', 'age', 'churn_date', 'city', 'tariff']]
df_user2
```

```python
df_new=df_new.merge(df_user2, how='left', on='user_id')
df_new
```

```python
#переведем лимит Мб в Гб и переименуем соответственно столбец
df_tariff['mb_per_month_included'] = df_tariff['mb_per_month_included']/1024
df_tariff = df_tariff.rename(columns={'mb_per_month_included': 'gb_per_month_included'})
```

```python
df_new=df_new.merge(df_tariff, how='left', left_on=['tariff'], right_on=['tariff_name'])
df_new
```

Теперь, имея ДФ со всеми необходимыми нам данными, для расчета помесячной выручки с каждого пользователя нам необходимо вычесть бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика, а остаток умножить на значение из тарифного плана и прибавить абонентскую плату, соответствующую тарифному плану. Для этого напишем функцию.

```python
#создаем локальные переменные фиксированной платы и по расчету дополнительной платы за перерасход 
#лимита сообщений/интернет-траффика/минут соответственно и суммируем их к стоимости тарифа, если есть превышение
def income(row):
    income = row['rub_monthly_fee']
    income_msg = (row['amount_msg'] - row['messages_included']) * row['rub_per_message']
    income_gb = (row['gb_used'] - row['gb_per_month_included']) * row['rub_per_gb']
    income_call = (row['total_duration'] - row['minutes_included']) * row['rub_per_minute']
    if row['amount_msg'] > row['messages_included']:
        income += income_msg
    if row['gb_used'] > row['gb_per_month_included']:
        income += income_gb
    if row['total_duration'] > row['minutes_included']:
        income += income_call
    return income
```

```python
#применяем нашу функцию и смотрим на результат
df_new['income'] = df_new.apply(income, axis=1)
df_new.head(10)
```

## Проверка гипотез

### Различается ли средняя выручка пользователей тарифов «Ультра» и «Смарт»

Для проверки этой гипотезы сформируем нулевую и альтернативную гипотезу.  
**Нулевая H0:** средняя выручка пользователей тарифа «Ультра» равна средней выручке пользователей «Смарт»  
**Альтернативная H1:** средняя выручка пользователей тарифа «Ультра» не равна средней выручке пользователей «Смарт»

```python
# для начала посчитаем среднее нашей выборки
display(smart.income.mean())
ultra.income.mean()
```

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(smart.income, ultra.income, equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Полученное очень маленькое значение p-value говорит о том, что вероятность получить такое или большее различие точно не случайно. Это явно слишком маленькая вероятность, чтобы делать вывод о тождественности между выручками двух тарифов.
Мы использовали дополнительный параметр equal_var = False, потому что у нас достаточно большие, но неодинаковые по размеру выборки, к тому же в каждой измерен разный тип пользователей.

### Различается ли средняя выручка пользователей из Москвы и пользователей из других регионов

Для проверки этой гипотезы сформируем нулевую и альтернативную гипотезу.  
**Нулевая H0:** средняя выручка пользователей из Москвы равна средней выручке пользователей из других регионов  
**Альтернативная H1:** средняя выручка пользователей из Москвы не равна средней выручке пользователей из других регионов 

Для начала нам необходимо разделить наш ДФ на два по городам пользователей.

```python
msk = df_new.query('city == "Москва"')
rgn = df_new.query('city != "Москва"')
```

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(msk.income, rgn.income)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно почти на 50%. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между пользователями из Москвы и других регионов.
Мы не использовали дополнительный параметр equal_var = False, потому что у нас достаточно большие выборки, к тому же в обеих измерена одна и та же переменная — нет оснований полагать, что разброс будет сильно отличаться.

- средняя выручка пользователей из Москвы равна средней выручке пользователей из других регионов
